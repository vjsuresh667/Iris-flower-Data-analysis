```python
import pandas as pd

# Load the dataset
df = pd.read_csv("Iris.csv")

# Drop the 'Id' column as it is not a feature
df = df.drop('Id', axis=1)

# Display the first few rows and column information
print(df.head())
print(df.info())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# Separate features (X) and target (y)
X = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
y = df['Species']

# Encode the target variable
le = LabelEncoder()
y_encoded = le.fit_transform(y)
target_names = le.classes_

# Normalize the features
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

# Split the data into training and testing sets (70/30 split)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled_df, y_encoded, test_size=0.3, random_state=42, stratify=y_encoded
)

# Experiment with different K values and evaluate
k_values = [1, 3, 5, 7, 9, 11]
results = {}

for k in k_values:
    # Train the model
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)

    # Predict on the test set
    y_pred = knn.predict(X_test)

    # Evaluate
    accuracy = accuracy_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)

    results[k] = {
        'accuracy': accuracy,
        'confusion_matrix': cm
    }

# Find the best K based on accuracy
best_k = max(results, key=lambda k: results[k]['accuracy'])
best_accuracy = results[best_k]['accuracy']
best_cm = results[best_k]['confusion_matrix']

print(f"--- KNN Experiment Results (K={k_values}) ---")
for k, res in results.items():
    print(f"K={k}: Accuracy = {res['accuracy']:.4f}")

print("\n--- Best Model Evaluation ---")
print(f"Best K: {best_k} with Accuracy: {best_accuracy:.4f}")
print("\nConfusion Matrix for Best K:")
print(best_cm)
print(f"Target Names (Labels): {target_names}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Step 5: Visualize Decision Boundaries ---
# 1. Select the two most discriminating features: Petal Length and Petal Width.
feature_names = ['PetalLengthCm', 'PetalWidthCm']
X_train_viz = X_train[feature_names]
X_test_viz = X_test[feature_names]

# 2. Retrain the classifier with K=5 using only these two features
best_k = 5 # Choosing K=5 as a stable choice among the equally best performing Ks
knn_viz = KNeighborsClassifier(n_neighbors=best_k)
knn_viz.fit(X_train_viz, y_train)

# 3. Create a meshgrid to plot the decision boundary
x_min, x_max = X_scaled_df[feature_names[0]].min() - 0.1, X_scaled_df[feature_names[0]].max() + 0.1
y_min, y_max = X_scaled_df[feature_names[1]].min() - 0.1, X_scaled_df[feature_names[1]].max() + 0.1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                     np.linspace(y_min, y_max, 100))

# Predict on the meshgrid points
Z = knn_viz.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# 4. Plot the decision boundary and the test data points
plt.figure(figsize=(10, 6))

# Plot decision boundary (colored background)
cmap_light = plt.cm.get_cmap('Spectral_r', 3)
plt.pcolormesh(xx, yy, Z, cmap=cmap_light, shading='auto')

# Plot the test data points
cmap_bold = plt.cm.get_cmap('Spectral', 3)
scatter = plt.scatter(X_test_viz.iloc[:, 0], X_test_viz.iloc[:, 1], c=y_test, cmap=cmap_bold,
                      edgecolor='k', s=60)

# Add labels and title
plt.xlabel(f'Normalized {feature_names[0]}')
plt.ylabel(f'Normalized {feature_names[1]}')
plt.title(f'KNN Decision Boundary (K={best_k}) with Test Data\nFeatures: Petal Length and Petal Width')

# Create a legend
legend1 = plt.legend(*scatter.legend_elements(),
                    loc="lower right", title="Classes", labels=target_names)
plt.gca().add_artist(legend1)

# Save the plot
plt.savefig('knn_decision_boundary.png')

print("Decision boundary visualization saved as 'knn_decision_boundary.png'.")
```
